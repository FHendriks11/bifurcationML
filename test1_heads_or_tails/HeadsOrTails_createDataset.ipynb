```python
import numpy as np
import torch
import matplotlib.pyplot as plt
import pickle
```

# Create data set

```python
# pick 10 integers between 1 and 2
rng = np.random.default_rng()
```

```python
N = 1000
rng = np.random.default_rng()
x = rng.uniform(-100, 100, N)
sign = rng.integers(0, 2, N)*2-1
y = x*sign

y = np.stack([y, -y], axis=1)
```

```python
# convert to torch tensors and scale
x = torch.tensor(x).float()
y = torch.tensor(y).float()
x_m = torch.mean(x).float().item()
y_m = torch.mean(y).float().item()
x_std = torch.std(x).float().item()
y_std = torch.std(y).float().item()
x_scaled = (x - x_m) / x_std
y_scaled = (y - y_m) / y_std
```

```python
plt.scatter(x.numpy(), y[:, 0].numpy(), s=1)
plt.xlabel('bet (€)')
plt.ylabel('winnings (€)')
plt.gca().set_aspect('equal')
plt.show()
```

```python
train_frac = 0.8
N_train = int(train_frac*len(x))
data_tr = torch.cat((x_scaled[:N_train].unsqueeze(-1), y_scaled[:N_train]), dim=1)
data_te = torch.cat((x_scaled[N_train:].unsqueeze(-1), y_scaled[N_train:]), dim=1)
data_tr.shape, data_te.shape
```

```python
# save data
with open('../data/HeadsOrTails_data.pkl', 'wb') as f:
    pickle.dump({'data_tr': data_tr,
                 'data_te': data_te,
                    'x_m': x_m,
                    'y_m': y_m,
                    'x_std': x_std,
                    'y_std': y_std
                 }, f)
```
